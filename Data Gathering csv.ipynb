## 1. Opening a local csv file

### CONVERTORS 
#### convert  one name into the other name for easy use replace function of excel

```python
pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
#column name and function name in dictionary
```
